# [1] Categorical Analysis

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(12,13)})
p=1
for i in ['Ship Mode','Segment','Region','Category','Sub-Category']:
    plt.subplot(3,2,p)
    if i=='Sub-Category':
        ax=sns.countplot(x=i,data=df)
        for i in ax.containers:
            ax.bar_label(i)
        plt.xticks(rotation=90)
    else:
        ax=sns.countplot(x=i,data=df)
        for i in ax.containers:
            ax.bar_label(i)
    p+=1
```

## Category and Subcategory wise sales

```python
ax=df.groupby('Category')['Sales'].mean().plot(kind='bar',label='Mean Sales',figsize=(7,5))
for i in ax.containers:
    ax.bar_label(i,)
plt.legend()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(x='Sub-Category',y='Sales',data=df,errorbar=None)
for i in ax.containers:
    ax.bar_label(i,fontsize=9)
plt.xticks(rotation=90)
plt.show()
```

```python
ax=sns.catplot(y='Sub-Category',x='Sales',data=df,kind='bar',errorbar=None,col='Region',col_wrap=2,
           height=6,aspect=2)
```

```python
pd.crosstab(index=df['Segment'],columns=df['Ship Mode'],margins=True)
```

```python
a=df.groupby('Sub-Category')[['Sales','Profit']].mean().reset_index()
a.head()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
ax1=plt.subplot(1,1,1)
sns.barplot(x='Sub-Category',y='Sales',data=a,label='Mean Sales')
plt.xticks(rotation=90)
plt.legend(loc='upper left')
ax2=ax1.twinx()
sns.lineplot(x='Sub-Category',y='Profit',data=a,linestyle='--', linewidth =3.5,
            marker ='o',markerfacecolor='black',color='r',markersize=9,label='Mean Profit')
ax2.grid(False)
```
